```python
import optuna
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Ruta a la base de datos SQLite3 de Optuna
storage = "sqlite:///op-class2.sqlite3"

# Nombres de los estudios que deseas comparar
study_names = ["lgbm kappa sin monto corregido", 
               "lgbm kappa sin monto textmining",
               "lgbm kappa sin monto textmining custom"]

# Cargar los estudios
studies = [optuna.load_study(study_name=name, storage=storage) for name in study_names]

# Graficar la historia de optimización
fig = optuna.visualization.plot_optimization_history(studies)

# Personalizar el tamaño de los puntos
for trace in fig.data:
    trace.marker.size = 6  # Ajusta este valor para hacer los puntos más pequeños (o más grandes)
fig.update_yaxes( range=[0.94, 1.02])
# Cambiar el título del gráfico
fig.update_layout(title="Historial de optimización score KAPPA")
# Mostrar el gráfico personalizado
fig.show()
```

Se observa que los resultados con text mining son superiores
